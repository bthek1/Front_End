- `class-variance-authority` + `clsx` + `tailwind-merge` — utility class management (shadcn/ui pattern)

- Playwright (E2E)

**Dev Tooling**
- ESLint 10 + `typescript-eslint` + `eslint-plugin-react-hooks`
